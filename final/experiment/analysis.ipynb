```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
result_file = open('output/ridge_regression_alpha100_submission.csv', 'r')
result_file.readline()
lines = []
for (i, line) in enumerate(result_file):
    lines.append((i,line))
```

```python
select = lines[2] # change the index to see the result
i, line = select[0], select[1]

# print description
with open('data/descriptions_test/'+str(i)+'.txt', 'r') as f:
    print('description')
    print(f.read())

top20 = line.strip().split(',')[1].split()
# print top 5 image and tag for analyzing
for j in range(5): 
    with open('data/tags_test/'+str(top20[j].split('.')[0])+'.txt', 'r') as f:
        print('tag')
        print(f.read())
    img=mpimg.imread('data/images_test/'+str(top20[j]))
    plt.figure()
    plt.imshow(img)
result_file.close()
```
