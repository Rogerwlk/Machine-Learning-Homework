```python
import numpy as np
from scipy.spatial import distance
from IPython.display import Image
```

```python
## read feature vectors 
f = open('./data/features_train/features_resnet1000_train.csv', 'r')

i = 0;
feature_vector_train = []
for line in f:
    tmp = line.strip().split(',')
    index = int(tmp[0][13:-4])
    vector = [float(i) for i in tmp[1:]]
    feature_vector_train.append([index,vector])
    #print(feature_vector_train[-1])
    i=i+1
    if i>30000:
        break
f.close()

# sort feature_vector_train accroding to index
def take_index(elem):
    return[elem[0]]
feature_vector_train.sort(key=take_index)
```

```python
##Specify the index of the image that you want to search
image_index_baseline = 403

##calculate the euclidean distance between the specified image and all the images
feature_dist = []
for i in range(len(feature_vector_train)):
    dist = distance.euclidean(feature_vector_train[image_index_baseline][1],feature_vector_train[i][1])
    #i,j,index
    feature_dist.append([feature_vector_train[image_index_baseline][0],feature_vector_train[i][0],dist])
        
# sort according to the distance
def take_distance(elem):
    return elem[2]
feature_dist.sort(key=take_distance)

##Display the specified image 
"image_index",feature_dist[0][1],"feature_distance",feature_dist[0][2]
Image(filename='./data/images_train/'+ str(image_index_baseline)+ '.jpg')
```
