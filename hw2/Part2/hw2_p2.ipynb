```python
# question b, c
from nltk.stem.porter import PorterStemmer
from nltk import word_tokenize
from collections import Counter
from sklearn.linear_model import LogisticRegressionCV
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import LassoCV

stop_words = set()
f = open('data/stops.txt', 'r')
for line in f:
    stop_words.add(line.strip())
f.close()

stemmer = PorterStemmer()
train_comments = []
train_ratings = []
test_comments = []
test_ratings = []

def read_file(fname, train_num = 400):
    good = bad = 0
    f = open(fname, 'r')
    for line in f:
        temp = line.strip().lower()
        comment, rate = temp.split('\t')
        rate = int(rate)
        comment = comment.replace('.', ' ')
        li = word_tokenize(comment)
        c = Counter([stemmer.stem(x) for x in li if x not in stop_words])
        if rate == 1:
            if good < train_num:
                train_ratings.append(rate)
                train_comments.append(c)
            else:
                test_ratings.append(rate)
                test_comments.append(c)
            good += 1
        else:
            if bad < train_num:
                train_ratings.append(rate)
                train_comments.append(c)
            else:
                test_ratings.append(rate)
                test_comments.append(c)
            bad += 1
    print(good,bad)
    
    f.close()
read_file('data/amazon_cells_labelled.txt')
read_file('data/yelp_labelled.txt')
read_file('data/imdb_labelled.txt')
```

```python
# question d
d = {}
bag = set()

for c in train_comments:
    for w in c.keys():
        bag.add(w)
for (i, w) in enumerate(bag):
    d[w] = i
    d[i] = w
for key in bag:
    print(train_comments[0][key], end=' ')
print()
for key in bag:
    print(train_comments[2399][key], end=' ')
print()
```

```python
# question e
def normalize(data):
    for c in data:
        s = sum(c.values())
        for k in c.keys():
            c[k] /= s

normalize(train_comments)
normalize(test_comments)
```

```python
# question f
train_pos = Counter()
train_neg = Counter()
for i in range(len(train_comments)):
    if train_ratings[i] == 1:
        train_pos += train_comments[i]
    else:
        train_neg += train_comments[i]
confusion_mx = [[0, 0], [0, 0]]
alpha = 0.1
correct = 0
for i in range(len(test_comments)):
    pos = 1
    neg = 1
    for (key, val) in test_comments[i].items():
        pos *= (train_pos[key] + alpha) ** val
        neg *= (train_neg[key] + alpha) ** val
    judge = int(pos > neg)
    if judge == test_ratings[i]:
        correct += 1
        if judge == 1:
            confusion_mx[0][0] += 1
        else:
            confusion_mx[1][1] += 1
    else:
        if judge == 1:
            confusion_mx[1][0] += 1
        else:
            confusion_mx[0][1] += 1
for row in confusion_mx:
    print(row)
print()
print('accuracy:', correct / len(test_comments))
```

```python
# question g
train_x = [[0 for x in range(len(bag))] for y in range(len(train_comments))]
for (i, c) in enumerate(train_comments):
    for (k, v) in c.items():
        if k in d:
            train_x[i][d[k]] = v
train_y = train_ratings

test_x = [[0 for x in range(len(bag))] for y in range(len(test_comments))]
for (i, c) in enumerate(test_comments):
    for (k, v) in c.items():
        if k in d:
            test_x[i][d[k]] = v
test_y = test_ratings
```

```python
ridge_reg = LogisticRegressionCV(cv=10,penalty='l2').fit(train_x, train_y)
```

```python
print('ridge_reg score: ', ridge_reg.score(test_x, test_y))
ridge_coef = [(ridge_reg.coef_[0][i], i) for i in range(len(ridge_reg.coef_[0]))]
order = sorted(ridge_coef)
for c in order[:5]:
    print(c[0], d[c[1]])
print()
for c in order[-5:]:
    print(c[0], d[c[1]])
print()
```

```python
print('lasso_reg score: ', lasso_reg.score(test_x, test_y))
lasso_coef = [(lasso_reg.coef_[0][i], i) for i in range(len(lasso_reg.coef_[0]))]
order = sorted(lasso_coef)
for c in order[:5]:
    print(c[0], d[c[1]])
print()
for c in order[-5:]:
    print(c[0], d[c[1]])
print()
```

```python
# question h
train_ratings_bigram = []
train_comments_bigram = []
test_ratings_bigram = []
test_comments_bigram = []

def read_bigram(fname, train_num = 400):
    good = bad = 0
    f = open(fname, 'r')
    for line in f:
        temp = line.strip().lower()
        comment, rate = temp.split('\t')
        rate = int(rate)
        comment = comment.replace('.', ' ')
        li = word_tokenize(comment)
        c = Counter([stemmer.stem(x) for x in li if x not in stop_words])
#         li = [x for x in li if x not in stop_words]
        for i in range(1, len(li)):
#             if li[i - 1] not in stop_words and li[i] not in stop_words:
            c[li[i - 1]+' '+li[i]] += 1
        if rate == 1:
            if good < train_num:
                train_ratings_bigram.append(rate)
                train_comments_bigram.append(c)
            else:
                test_ratings_bigram.append(rate)
                test_comments_bigram.append(c)
            good += 1
        else:
            if bad < train_num:
                train_ratings_bigram.append(rate)
                train_comments_bigram.append(c)
            else:
                test_ratings_bigram.append(rate)
                test_comments_bigram.append(c)
            bad += 1
    f.close()
read_bigram('data/amazon_cells_labelled.txt')
read_bigram('data/yelp_labelled.txt')
read_bigram('data/imdb_labelled.txt')
```
